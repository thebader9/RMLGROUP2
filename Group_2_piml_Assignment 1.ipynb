#### **Score test data**

#### Define input features as list

```python
x_names = ['term_360', 'conforming', 'debt_to_income_ratio_missing', 'loan_amount_std', 'loan_to_value_ratio_std', 'no_intro_rate_period_std',
           'intro_rate_period_std', 'property_value_std', 'income_std', 'debt_to_income_ratio_std']
```

**Score XGB2_v2 on test data**

```python
from google.colab import files           
uploaded = files.upload()
```

```python
test = pd.read_csv('hmda_test_preprocessed.csv')
```

```python
import pandas as pd
piml_XGB2 = pd.DataFrame(exp.experiment.registered_pipeline['XGB2'].predict_proba(test[x_names].to_numpy()), columns=['p0','phat'])
print(piml_XGB2.head())
piml_XGB2['phat'].to_csv('group2_piml_XGB2.csv', index=False)
files.download('group2_piml_XGB2.csv')
```

#### **Score FIGS on test data**

```python
piml_EBM = pd.DataFrame(exp.experiment.registered_pipeline['EBM'].predict_proba(test[x_names].to_numpy()), columns=['p0','phat'])
print(piml_EBM.head())
piml_EBM['phat'].to_csv('group2_piml_EBM.csv', index=False)
files.download('group2_piml_EBM.csv')
```

#### **Score ReLU-DNN on test data**

```python
piml_ReLUDNN = pd.DataFrame(exp.experiment.registered_pipeline['ReLU-DNN'].predict_proba(test[x_names].to_numpy()), columns=['p0','phat'])
print(piml_ReLUDNN.head())
piml_ReLUDNN['phat'].to_csv('group2_piml_reludnn.csv', index=False)
files.download('group2_piml_reludnn.csv')
```
